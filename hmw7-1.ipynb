```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from time import*
browser = webdriver.Chrome()
url = "http://rate.am/en/armenian-dram-exchange-rates/banks/non-cash"
browser.get(url)
data=browser.find_element_by_xpath('//*[@id="aspnetForm"]/div[3]/div[2]/div[1]/div[4]/div[3]/table/tbody/tr/td[1]/a')
data.click()
```

```python
year = browser.find_element_by_xpath('//*[@id="calBorder"]/span/table/tbody/tr[1]/td/table/tbody/tr/td[2]/select')
sy = Select(year)
sy.select_by_visible_text('2016')
```

```python
month = browser.find_element_by_xpath('//*[@id="calBorder"]/span/table/tbody/tr[1]/td/table/tbody/tr/td[1]/select')
sm = Select(month)
sm.select_by_visible_text('July')
```

```python
data1=browser.find_element_by_xpath('//*[@id="calBorder"]/span/table/tbody/tr[3]/td/table/tbody/tr[5]/td[3]')
data1.click()
```

```python
data2=browser.find_element_by_xpath('//*[@id="ctl00_Content_RB_dtpick1_btnOk"]')
data2.click()
```

```python
import pandas as pd
import csv
opt = pd.read_html(browser.find_element_by_class_name("rb").get_attribute('outerHTML'))
opt= opt[0]
opt = opt[2:19]
del opt[0]
del opt[2]
del opt[3]
opt.columns = ["name", "date", "USD_buy", "USD_sell", "EUR_buy", "EUR_sell", "RUB_buy", "RUB_sell", "GBP_buy", "GBP_sell"]
from unicodecsv import*
with open('datadata.csv', 'w') as f:
            opt.to_csv(f, header=True, index=False, encoding='utf-8')
```
